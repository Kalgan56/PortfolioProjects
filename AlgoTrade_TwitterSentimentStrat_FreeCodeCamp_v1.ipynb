# Twitter Sentiment Investing strategy
## 1. Load Twitter sentiment data
- Load the twitter sentiment dataset, set the index, calculate the engagement ratio and filter out stocks with no significant twitter activity

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import yfinance as yf
import os
plt.style.use('ggplot')

sentiment_df=pd.read_csv('sentiment_data.csv')
sentiment_df
```

```python
# set date to datetime format
sentiment_df['date']=pd.to_datetime(sentiment_df['date'])

sentiment_df=sentiment_df.set_index(['date','symbol'])
sentiment_df
```

```python
# calculate engagement ratio to try and filter out bot activity
# if ratio of likes to comments is high then probably lots of bots
sentiment_df['engagement_ratio']=sentiment_df['twitterComments']/sentiment_df['twitterLikes']
sentiment_df
```

```python
# filter out stocks with not a lot of comments or likes
sentiment_df_2=sentiment_df[(sentiment_df['twitterLikes']>20)&(sentiment_df['twitterComments']>10)]
sentiment_df_2
```

## 2. Aggregate monthly and calculate average sentiment for the month
- Aggregate on the monthly level and calculate average monthly metric, for the one we choose

```python
# basically make monthly groups for each symbol and then take the mean of the engagement ratio for the whole month
aggregated_df=(sentiment_df_2.reset_index('symbol').groupby([pd.Grouper(freq='M'),'symbol'])[['engagement_ratio']].mean())

# rank the symbols based on the engagement ratio. high engagement ratio leads to a higher rank, low engagement ratio is a lower rank since it's probably bots
aggregated_df['rank']=(aggregated_df.groupby(level=0)['engagement_ratio'].transform(lambda x: x.rank(ascending=False)))

aggregated_df
```

## 5. Download fresh stock prices for only select/shortlisted stocks

```python
stocks_list=sentiment_df_2.index.get_level_values('symbol').unique().tolist()

prices_df=yf.download(tickers=stocks_list,
                      start='2021-01-01',
                      end='2023-03-01')
```

## 6. Calculate portfolio returns with monthly rebalancing

```python
returns_df=np.log(prices_df['Adj Close']).diff()
returns_df2=returns_df.drop(returns_df.index.values[0])
returns_df2=returns_df2.drop(columns=['ATVI'])
portfolio_df=pd.DataFrame()

for start_date in fixed_dates.keys():
    end_date= (pd.to_datetime(start_date)+pd.offsets.MonthEnd()).strftime('%Y-%m-%d')

    cols=fixed_dates[start_date]
    
    temp_df= returns_df2[start_date:end_date][cols].mean(axis=1).to_frame('portfolio_return') # equally weighted returns
    
    portfolio_df=pd.concat([portfolio_df,temp_df],axis=0)
portfolio_df
```

## 7. Download NASDAQ/QQQ prices and calculate returns to compare to our strategy

```python
qqq_df=yf.download(tickers='QQQ',
                   start='2021-01-08',
                   end='2023-03-01')
qqq_ret=np.log(qqq_df['Adj Close']).diff().to_frame('nasdaq_return')

portfolio_df=portfolio_df.merge(qqq_ret,
                                left_index=True,
                                right_index=True)
portfolio_df
```

```python
import matplotlib.ticker as mtick
portfolios_cumulative_return=np.exp(np.log1p(portfolio_df).cumsum().sub(1))

portfolios_cumulative_return.plot(figsize=(16,6))

plt.title('Twitter Engagement Ratio Strategy Return Over Time')

plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(1))

plt.ylabel('Return')

plt.show()
```
